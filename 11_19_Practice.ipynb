```python
import geopandas as gpd
import os
import fiona
import matplotlib.pyplot as plt
```

```python
gdb_path=r"C:\Users\bjetra.UOM\Downloads\geoportal.gdb\geoportal.gdb"
layers=fiona.listlayers(gdb_path)
```

```python
for layer in layers:
    print(f"layer name is {layer}")
    gdf=gpd.read_file(gdb_path, layer=layer)
    print(f"epsg crs is {gdf.crs}")
    print(f"number of rows is {len(gdf)}")
    print(f"geometry type is {gdf.geom_type.unique()}")
    print(f"colums are{list(gdf.columns)}\n")
    gdf.plot()
    fig, ax=plt.subplots(figsize=(8,6))
    gdf.plot(ax=ax)
    ax.set_title(f"layer: {layer}")
    plt.show()
```

```python
def explore_gdb(gdb_path):
    layers=fiona.listlayers(gdb_path)
    for layer in layers:
        print(f"layer name is {layer}")
        gdf=gpd.read_file(gdb_path, layer=layer)
        print(f"epsg crs is {gdf.crs}")
        print(f"number of rows is {len(gdf)}")
        print(f"geometry type is {gdf.geom_type.unique()}")
        print(f"colums are{list(gdf.columns)}\n")
        
        fig, ax=plt.subplots(figsize=(8,6))
        gdf.plot(ax=ax)
        ax.set_title(f"layer: {layer}")
        plt.show()
```

```python
explore_gdb(r"C:\Users\bjetra.UOM\Downloads\geoportal.gdb\geoportal.gdb")
```
